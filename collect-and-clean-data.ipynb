```python
# do before above: clean up datasets
import pandas as pd
import os

# Initialize an empty dataframe with the required columns
dfNew = pd.DataFrame(columns=["text"])

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        dataset = os.path.join(dirname, filename)
        try:
            df = pd.read_csv(dataset)
            # Drop NA values for each specified column and append to the dfNew dataframe
            for col in ["text", "Text", "body", "content", "Title of the Episode", "Description", "description", "line", "comment", "Title", "Article","headline_text", "message"]:
                if col in df.columns:
                    df_col = df.dropna(subset=[col])[col]
                    dfNew = dfNew.append(pd.DataFrame({ "text": df_col }))
        except pd.errors.EmptyDataError:
            print("Empty CSV file. Skip. Name: " + dataset)
        except pd.errors.ParserError:
            print("Invalid data format. Skip. Name: " + dataset)
        except:
            print("Unexpected error. Skip. Name: " + dataset)
dfNew.to_csv("data2.csv", index=False)
```
